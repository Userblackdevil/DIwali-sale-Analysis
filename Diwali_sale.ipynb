```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()#checking the values 
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)#drop empty or unwanted columns 
```

```python
df.info()
```

```python
pd.isnull(df).sum()#checking the null values
```

```python
df.dropna(inplace=True)# Removing the values
```

```python
df.shape
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})#  It is used to change the columns name
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_Age
```

```python
sns.barplot(data=sales_Age,x='Age Group',y='Amount')
```

```python
df.columns
```

```python
sales_orders=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
ax=sns.barplot(data=sales_orders,x='State',y='Orders')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(6,6)})
sns.countplot(data=df,x='Marital_Status')
```

```python
sales_marital=df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_marital,x='Marital_Status',y='Amount')
```
